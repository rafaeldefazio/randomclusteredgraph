# Importa bibliotecas

```python
from graph import CRG
from radius_diameter import EccentricityD
```

## Cria grafo

```python
A = CRG("A", N=300, M=10, K=10, PIN=0.8, POUT=0.5)
```

## Calculando excentricidade com Dijkstra

```python
ED = EccentricityD(A)
print("Raio: %d | Diâmetro: %d" % (ED.radius, ED.diameter))
```

## Compara com cálculo realizado com NetworkX e desenhando grafo

```python
import networkx as nx
nodes = nx.from_numpy_matrix(A.adj)

print("Raio: %d | Diâmetro: %d" % (nx.radius(nodes), nx.diameter(nodes)))
```

## Desenha grafo

```python
print("\n\nDesenhando grafo...\n\n")

cmaps = [
	            'Grey', 'Purple', 'Blue', 'Green', 'Orange', 'Red',
	            'Yellow','Grey', 'Purple', 'Blue', 'Green', 'Orange', 'Red','Grey', 'Purple', 'Blue', 'Green', 'Orange', 'Red',]
color_map = []

for i in range(0, A.M):
    for j in range(i * A.nodes_per_community,(i+1) * A.nodes_per_community):
        color_map.append(cmaps[i])
        
        
nx.draw(nodes, pos=nx.spring_layout(nodes), node_color = color_map,with_labels = False, node_size=30)
```
